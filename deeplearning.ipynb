```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist


(x_train, y_train), (x_test, y_test) = mnist.load_data() #データの読み込み
x_train, x_test = x_train / 255.0, x_test / 255.0 #正規化
model = keras.models.Sequential() #モデルの定義
model.add(layers.Flatten(input_shape=(28, 28))) #入力層
model.add(layers.Dense(128, activation='relu')) #中間層
model.add(layers.Dense(10, activation='softmax')) #出力層
model.compile(optimizer='adam', #最適化関数
                loss='sparse_categorical_crossentropy', #損失関数
                metrics=['accuracy']) #評価関数
model.fit(x_train, y_train, epochs=5,validation_data=(x_test, y_test)) #評価用データ
```

```python
import matplotlib.pyplot as plt # matplotをpltとしてインポート
import numpy as np # numpyをnpとしてインポート

plt.imshow(x_test[0], cmap='grays') # 画像を表示
plt.show() # 画像を表示
pre = model.predict(x_test) # テストデータを予測
index = np.argmax(pre[0]) # 最も確率の高い要素のインデックスを取得
print(f"この画像は{index}です")
```

```python
# and回路
x1 = [0, 1, 0, 1] # 入力1
x2 = [0, 0, 1, 1] # 入力2

def test(x1, x2): # and回路の定義
    if x1 == 1 and x2 == 1: # 入力1と入力2が1のとき
        return 1 # 1を返す
    else: # それ以外のとき
        return 0 # 0を返す
    
def disp_results(func): # 結果を表示する関数
    for i in range(4): # 4回繰り返す
        Y = func(x1[i], x2[i]) # and回路の実行
        print(f"{x1[i]}, {x2[i]} | {Y}") # 結果を表示

disp_results(test) # 結果を表示
```

```python
def and_test(x1, x2): # and回路のパーセプトロン
    w1, w2, theta = 0.5, 0.5, 0.8 # 重みと閾値の定義
    ans = x1 * w1 + x2 * w2 # 総入力の計算
    if ans > theta:
        return 1
    else:
        return 0
    
disp_results(and_test) # 結果を表示
```

```python
# orをパーセプトロンで実装
def or_test(x1, x2): # or回路のパーセプトロン
    w1, w2, theta = 0.5, 0.5, 0.2 # 重みと閾値の定義
    ans = x1 * w1 + x2 * w2 # 総入力の計算
    if ans > theta:
        return 1
    else:
        return 0
    
disp_results(or_test) # 結果を表示
```

```python
plot_perceptron(or_test, x1, x2) # or回路のパーセプトロンを表示
```

```python
# nandをパーセプトロンで実装
def nand_test(x1, x2): # nand回路のパーセプトロン
    w1, w2, theta = -0.5, -0.5, -0.8 # 重みと閾値の定義
    ans = x1 * w1 + x2 * w2 # 総入力の計算
    if ans > theta:
        return 1
    else:
        return 0
    
disp_results(nand_test) # 結果を表示
```

```python
plot_perceptron(nand_test, x1, x2) # nand回路のパーセプトロンを表示
```

```python
# xorをパーセプトロンで実装
def xor_test(x1, x2): # xor回路のパーセプトロン
    if or_test(x1, x2) > 0:
        s1 = 1
    else:
        s1 = 0
    if nand_test(x1, x2) > 0:
        s2 = 1
    else:
        s2 = 0
    ans = and_test(s1, s2)
    if ans > 0:
        return 1
    else:
        return 0
            
disp_results(xor_test) # 結果を表示
```

```python
plot_perceptron(xor_test, x1, x2) # xor回路のパーセプトロンを表示
```

```python
# 活性化関数
import numpy as np # numpyをnpとしてインポート
import matplotlib.pyplot as plt # matplotをpltとしてインポート

def step_func(X): # ステップ関数
    return np.where(X > 0, 1, 0) # ステップ関数の定義
x1 = np.linspace(-5, 5, 500) # x軸の値を生成
y1 = step_func(x1) # ステップ関数を計算

plt.plot(x1, y1) # ステップ関数をプロット
plt.legend(["step"], loc="best") # 凡例を表示
plt.yticks(np.arange(0, 1.2, step=0.5)) # y軸の目盛りを設定
plt.grid() # グリッドを表示
plt.show() # グラフを表示
```

```python
# シグモイド関数
def sigmoid_func(X): # シグモイド関数
    return 1 / (1 + np.exp(-X)) # シグモイド関数の定義
x2 = np.linspace(-5, 5) # x軸の値を生成
y2 = sigmoid_func(x2) # シグモイド関数を計算

plt.plot(x1, y1) # シグモイド関数をプロット
plt.plot(x2, y2) # ステップ関数をプロット   
plt.legend(["step", "sigmoid"], loc="best") # 凡例を表示
plt.yticks(np.arange(0, 1.2, step=0.5)) # y軸の目盛りを設定
plt.grid() # グリッドを表示
plt.show() # グラフを表示
```

```python
# ハイパーボリックタンジェント関数
def tanh_func(X): # ハイパーボリックタンジェント関数
    return np.tanh(X) # ハイパーボリックタンジェント関数の定義
x3 = np.linspace(-5, 5) # x軸の値を生成
y3 = tanh_func(x3) # ハイパーボリックタンジェント関数を計算

plt.plot(x1, y1) # ステップ関数をプロット
plt.plot(x2, y2) # シグモイド関数をプロット
plt.plot(x3, y3) # ハイパーボリックタンジェント関数をプロット
plt.legend(["step", "sigmoid", "tanh"], loc="best") # 凡例を表示
plt.yticks(np.arange(-1, 1.2, step=0.5)) # y軸の目盛りを設定
plt.grid() # グリッドを表示
plt.show() # グラフを表示
```

```python
# ReLU関数
def relu_func(X): # ReLU関数
    return np.where(X > 0, X, 0) # ReLU関数の定義
x4 = np.linspace(-5, 5) # x軸の値を生成
y4 = relu_func(x4) # ReLU関数を計算

plt.plot(x1, y1) # ステップ関数をプロット
plt.plot(x2, y2) # シグモイド関数をプロット
plt.plot(x3, y3) # ハイパーボリックタンジェント関数をプロット
plt.plot(x4, y4) # ReLU関数をプロット
plt.legend(["step", "sigmoid", "tanh", "ReLU"], loc="best") # 凡例を表示
plt.yticks(np.arange(-1, 1.2, step=0.5)) # y軸の目盛りを設定
plt.grid() # グリッドを表示
plt.show() # グラフを表示
```
